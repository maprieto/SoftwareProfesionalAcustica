# Propagación de ondas armónicas planas en un medio multicapa

Los medios multicapa son soluciones habituales para
muchos de los problemas que nos podemos encontrar en acústica. 
Problemas como el control de ruido y la reducción de vibraciones
requieren en la práctica paredes o revestimientos compuestos por 
una sucesión de diferentes materiales con los que controlar
la propagación de ondas acústicas. En la práctica, existen multitud 
de materiales con los que construir un medio multicapa y las posibles combinaciones
con las que nos podemos enfrentar pueden incluir cámaras de aire, 
materiales cerámicos, lanas de roca, láminas metálicas, velos porosos, etc.

De forma general, supondremos que un corte vertical de un medio multicapa formada por $n$ medios cuyas
interfaces están definidas por los planos $p=a_{j}$ para cada $j=1,\ldots,n-1$ está descrito como aparece en la siguiente figura:
<img src="images/multicapa.png" alt="Material multicapa" title="Material multicapa." width="600"/>

En cada una de las interfaces de acople entre dos medios (capas) consecutivas se denotará por $Z_{j}^{in}$ y $Z_{j}^{out}$ a la impedancia de superficie de entrada y salida asociada a la interfaz $j$, respectivamente.

En lo que sigue y para ilustrar de una forma simple los cálculos 
que se deben realizar para resolver
la propagación de ondas planas a través de un medio multicapa,
consideraremos el caso de ondas planas unidimensionales y 
supondremos que los desplazamientos de cada uno de los medios son suma de dos ondas
planas, una progresiva y otra regresiva. Las ecuaciones que 
satisfacen los campos de desplazamientos $u_j(p,t)$, para $j=1,2,\ldots,n$ son
\begin{equation*}
\frac{\partial^2 u_{j}}{\partial t^2}-c_{j}\frac{\partial^2u_{j}}{\partial p^2}=0\qquad
\text{ en el medio }j,
\end{equation*}
donde $c_{j}$ es la velocidad de propagación del medio $j$.

Puesto que no estamos interesados en la transición inicial del 
campo de desplazamientos, no daremos condiciones iniciales $u_j(p,0)$
o $\frac{\partial u_j}{\partial t}(p,0)$ para cada $j=1,2,\ldots,n$. 
En cambio, buscaremos soluciones armónicas donde sólo la primera componente
es no nula, es decir, el campo de desplazamientos $u_{j}$,
velocidades $v_{j}$ y presiones $\pi_j$ de la capa $j$-ésima son de la forma
\begin{align*}
&u_j(p,t)=\mathrm{Re}\left(e^{-i\omega t}U_j(p)\right)\mathbf{e}_{1},\\
&v_j(p,t)=\mathrm{Re}\left(e^{-i\omega t}V_j(p)\right)\mathbf{e}_{1},\\
&\pi_j(p,t)=\mathrm{Re}\left(e^{-i\omega t}\Pi_j(p)\right),
\end{align*}
para $j=1,2,\ldots,n$.

Si nuestro objetivo es simular el comportamiento acústico de la 
multicapa cuando la excitamos en el medio $1$
mediante una onda plana progresiva de amplitud compleja $F$,
podemos escribir las condiciones de contorno del problema como
$$
\left.\left(\frac{d U_{1}}{d p}+ik_{1}U_{1}\right)\right|_{p=a_{1}}=2ik_{1}Fe^{ik_{1}a_{1}},
$$
junto con la condición de radiación saliente
\begin{equation*}
\lim_{p\to+\infty}\left(\frac{d U_{n}}{d p}-ik_{n}U_{n}\right)=0,
\end{equation*}
donde $k_j$ es el número de onda del medio $j$. La ecuación
anterior asegura que no existen ondas regresivas
en el medio $n$ situado en el extremo derecho.

En consecuencia, en el dominio de la frecuencia, 
los desplazamientos complejos $U_{j}$ son solución del siguiente problema
en el que aún faltan las condiciones de transmisión:
Hallar los campos de desplazamientos complejos $U_j(p)$, 
para $j=1,2,\ldots,n$ tales que satisfacen las ecuaciones
\begin{equation*}
\left\{\begin{array}{rcll}
 k_{j}U_{j}+\frac{d^2U_{j}}{d p^2}&=&0&\text{ en el medio }j,\\
\lim_{p\to+\infty}\left(\frac{d U_{n}}{d p}-ik_{n}U_{n}\right)&=&0,&\\
\left.\left(\frac{d U_{1}}{d p}+ik_{1}U_{1}\right)\right|_{p=a_{1}}&=&2ik_{1}Fe^{ik_{1}a_{1}},&
\end{array}\right.
\end{equation*}

## Método de transferencia de matrices
Los campos de desplazamientos complejos $U_j$ se pueden escribir como
\begin{equation*}
U_j(p)=C_{2j-1}e^{-ik_{j}p}+C_{2j}e^{ik_{j}p},
\qquad j=1,2,\ldots,n,
\end{equation*}
donde las amplitudes $C_{2j-1}$, y $C_{2j}$ corresponden a 
las ondas regresivas y progresivas respectivamente, de cada medio.
Obviamente, las amplitudes de cada una de las ondas que forman 
parte del desplazamiento $U_{j}$, deben ser determinadas a partir de las condiciones
de transmisión de cada una de las interfaces de la multicapa.

Por otra parte, los diferentes tipos de interfaz considerados entre dos medios
(interfaz de contacto, placa rígida, panel viscoelástico, velo poroso, etc)
dan lugar a diferentes condiciones de transmisión, 
para cada interfaz $j=1,2,\ldots,n-1$, que se traducen en dos ecuaciones lineales respecto a las
amplitudes de los desplazamientos en cada una de las interfaces. 
Estas dos ecuaciones se pueden escribir matricialmente, en cada plano $p=a_j$, como
\begin{equation*}
A_j(C_{2j-1},C_{2j},C_{2j+1},C_{2j+2})^t=(0,0)^t,
\end{equation*}
siendo $A _j$ una matriz de dos filas y cuatro columnas, $A _j\in{\mathcal{M}}_{2\times 4}(\mathbb{C})$.
Cada matriz $A_j$ se puede descomponer en dos matrices $2\times 2$ para diferenciar las columnas que 
multiplican a las amplitudes del medio $j$ y $j+1$,
\begin{equation*}
A_j=\left(A _{j,j}\ A_{j,j+1}\right)
\qquad \mathrm{ con }A_{j,j+1},A_{j,j}\in\mathcal{M}_{2\times 2}(\mathbb{C}).
\end{equation*}
Entonces, si suponemos que la primera onda plana progresiva 
tiene una amplitud $F$, debemos resolver el siguiente sistema 
de $2n$ ecuaciones con $2n$ incógnitas:
\begin{equation*}
AC=F ,
\end{equation*}
donde $C=(C_1,\ldots,C_{2n})^t\in\mathbb{C}^{2n}$, 
$F=(F,0,\ldots,0)^t\in\mathbb{C}^{2n}$ y 
\begin{equation*}
A =
\begin{pmatrix}
A _{0,1}&                &                &                &                &                \\
A _{1,1}&A _{1,2}&                &                &                &                \\
                &A _{2,2}&A _{2,3}&                &                &                \\
            &                &A _{3,3}&A _{3,4}&                &                \\
                &                &                & \ddots         &                &                \\
                &                &                &                &A _{n-1,n-1}&A _{n-1,n}\\
                &                &                &                &                &A _{n,n}
\end{pmatrix}
\in\mathcal{M}_{2n\times 2n}(\mathbb{C}).
\end{equation*}
En este sistema  $A _{0,1}=(0\;1)\in\mathcal{M}_{1\times 2}$ 
y $A _{n,n}=(1\; 0)\in\mathcal{M}_{1\times 2}$.
La primera indica la existencia de una onda incidente progresiva 
en el primer  medio de amplitud $F$ y la última indica que no existen 
ondas regresivas en el medio $n$ (que no es más que una reinterpretación de la condición de radiación
de Sommerfeld).

## Impedancia de entrada y salida. Coeficientes de reflexión, absorción y transmisión
Al resolver el sistema lineal anterior se obtiene 
el vector $C$ que tiene por componentes las amplitudes de las ondas planas
que son solución del problema de propagación acústica a través del medio multicapa. 
A partir de estas amplitudes podemos calcular todas las cantidades físicas de interés
(intensidad, potencia,etc.) y los coeficientes de reflexión y transmisión entre cada capa.

Teniendo en cuenta la solución del problema lineal anterior, la solución del problema de propagación unidimensional que estamos
considerando, esto es, tanto la presión $\Pi_{j}$ como la velocidad $V_{j}$ y el desplazamiento $U_{j}$ 
en el medio $j$ se pueden expresar como suma de ondas progresivas y regresivas
\begin{align*}
\Pi_{j}(p)&=\Pi_{\mathrm{pro},j}(p)+\Pi_{\mathrm{reg},j}(p),\\
V_{j}(p)&=V_{\mathrm{pro},j}(p)+V_{\mathrm{reg},j}(p),\\
U_{j}(p)&=U_{\mathrm{pro},j}(p)+V_{\mathrm{reg},j}(p),
\end{align*}
donde
\begin{align*}
\Pi_{\mathrm{pro},j}(p)&=-i\omega Z_{j}C_{2j-1}e^{-ik_{j}p},\qquad
\Pi_{\mathrm{reg},j}(p)=i\omega Z_{j}C_{2j}e^{ik_{j}p},\\
V_{\mathrm{pro},j}(p)&=-i\omega U_{\mathrm{pro},j}(p)=-i\omega C_{2j-1}e^{-ik_{j}p},\qquad
V_{\mathrm{reg},j}(p)=-i\omega U_{\mathrm{reg},j}(p)=-i\omega C_{2j}e^{ik_{j}p},
\end{align*}
y se satisface las relaciones entre presiones y velocidades siguientes:
\begin{align*}
\Pi_{\mathrm{pro},j}(p)&=Z_{j}V _{\mathrm{pro},j}(p)=-i\omega Z_{j} U_{\mathrm{pro},j}(p),\\
\Pi_{\mathrm{reg},j}(p)&=-Z_{j}V _{\mathrm{reg},j}(p)=i\omega Z_{j} U_{\mathrm{reg},j}(p)
\end{align*}
donde $Z_{j}$ denota la impedancia característica del medio $j$.

En consecuencia, si utilizamos este hecho y la definición de intensidad acústica (producto de presión por velocidad),
la intensidad acústica de la onda plana progresiva
$i_{\mathrm{pro},j}=\mathrm{Re}(e^{-i\omega t}I_j\mathbf{e}_1)$ en el medio $j$ se puede escribir como
\begin{align*}
I_j(p)&=(i_{\mathrm{pro},j})_{\text{av}}(p)\cdot\mathbf{e}_1
=-\frac{\omega}{2}\mathrm{Im}\left(\Pi_{\mathrm{pro},j}(p)
\overline{U }_{\mathrm{pro},j}(p)\right) \\
&=\frac{\omega^2}{2}\mathrm{Re}\left(Z_j\left|U_{\mathrm{pro},j}(p)\right|^2\right)
=\frac{\omega^2\mathrm{Re}(Z_j)}{2}|C_{2j}|^2|e^{ik_{j}p}|^2,\ j=1,2,\ldots,n.
\end{align*}
Si el medio $j$ es un fluido compresible, tanto la impedancia $Z_j$ 
como el número de onda $k_{j}=\frac{\omega}{c_{j}}$
son reales y por tanto,
\begin{equation*}
I_j(p)=\frac{\omega^2 Z_j}{2}|C_{2j}|^2.
\end{equation*}
Volviendo al caso general y teniendo en cuenta la expresión de la intesidad, 
el coeficiente de transmisión entre los medios $j$ y $j+1$
se define como
\begin{equation*} 
\tau_j=\frac{I_{j+1}(a_j)}{I_j(a_j)}=
\frac{|C_{2j+2}|^2|e^{ik_{j+1}a_j}|^2\mathrm{Re}(Z_{j+1})}
{|C_{2j}|^2|e^{ik_{j}a_j}|^2\mathrm{Re}(Z_{j})},\ j=1,2,\ldots,n-1.
\end{equation*}
También puede ser de interés saber cuál es el coeficiente de transmisión entre el
medio del que parte la onda que excita la multicapa y el medio $n$,
definido como $\tau=\tau_1\ldots\tau_{n-1}$.

Si suponemos que todos los medios son no disipativos obtenemos
\begin{equation*}
\tau=\tau_1\cdots\tau_{n-1}=\prod_{j=1}^{n-1}\frac{I_{j+1}(a_j)}{I_j(a_j)}
=\prod_{j=1}^{n-1}\frac{Z_{j+1}|C_{2j+2}|^2}{Z_j|C_{2j}|^2}
= \frac{Z_{n}|C_{2n}|^2}{Z_{1}|C_{2}|^2},
\end{equation*}
lo que da una valoración de la pérdida de energía 
que sufrió la onda al viajar a través de la multicapa.

De la misma forma, también podemos escribir el valor eficaz 
de presión sonora en cada medio en función de las amplitudes
de las ondas planas, para cada $j=1,2,\ldots,n$, como
\begin{align*}
(\pi_{\mathrm{rms},j})(p)&=\frac{1}{\sqrt{2}}|\Pi_j(p)|=\frac{\omega}{\sqrt{2}}|Z_j|
|-C_{2j-1}e^{-ik_{j}p}+C_{2j}e^{ik_{j}p}|.
\end{align*}

Asociadas a estas medidas de valoración del sonido, es usual 
definir la impedancia de entrada y salida en la interfaz $j$:

### Impedancias de entrada y salida
Dado un medio multicapa formado por $n$ medios,
la **impedancia de entrada** en la interfaz $j$-ésima de la multicapa
se define por
\begin{equation*}
Z^{\mathrm{in}}_{j}=\frac{\Pi_j(a_j)}{V _j(a_j)},\ j=1,2,\ldots,n-1.
\end{equation*}
mientras que la **impedancia de salida** en la interfaz $j$-ésima de la multicapa viene
dada por la expresión
\begin{equation*}
Z^{\mathrm{out}}_{j}=\frac{\Pi_{j+1}(a_j)}{V _{j+1}(a_j)},\ j=1,2,\ldots,n-1.
\end{equation*}

Estas impedancias pueden escribirse como
\begin{align*}
Z^{\mathrm{in}}_{j}&=Z_j\frac{-C_{2j-1}e^{-ik_{j}a_j}+C_{2j}e^{ik_{j}a_j}}
{C_{2j-1}e^{-ik_{j}a_j}+C_{2j}e^{ik_{j}a_j}}, \\
Z^{\mathrm{out}}_{j}&=Z_{j+1}\frac{-C_{2j+1}e^{-ik_{j+1}a_j}+C_{2j+2}e^{ik_{j+1}a_j}}
{C_{2j+1}e^{-ik_{j+1}a_j}+C_{2j+2}e^{ik_{j+1}a_j}}.
\end{align*}
Si la condición entre los dos medios es de contacto, se obtiene inmediatamente
$$
Z^{\mathrm{in}}_{j}=\frac{\Pi_j(a_j)}{V_{j}(a_j)}
=\frac{\Pi_{j+1}(a_j)}{V_{j}(a_{j+1})}=Z^{\mathrm{out}}_{j}.
$$

Es habitual definir la impedancia de entrada del medio multicapa como $Z_I=Z^{\mathrm{in}}_{1}$
y la impedancia de salida del medio multicapa como $Z_T=Z^{\mathrm{out}}_{n-1}$.

### Coeficientes de reflexión, absorción y transmisión

El nivel de presión sonora, denotado por SPL o bien por $L_{\pi}$, se define como el cociente entre el
valor eficaz de la presión sonora $\pi_{\text{rms}}$ 
y el valor de referencia o umbral de presión $\pi_{\text{ref}}=2\times 10^{-5}$ Pa,
\begin{equation*}
\text{SPL}=20\log_{10}\left(\frac{\pi_{\text{rms}}}{\pi_{\text{ref}}}\right).
\end{equation*}

En el régimen armónico, los coeficientes de reflexión y transmisión 
asociados a la presión son los números complejos $R$ y $T$ definidos 
respectivamente como
$$
\begin{array}{ccc}
R&=&\frac{\Pi_{\text{refl}}}{\Pi_{\text{inc}}},\\
T&=&\frac{\Pi_{\text{tran}}}{\Pi_{\text{inc}}},
\end{array}
$$
donde $\Pi_{\text{inc}}$, $\Pi_{\text{refl}}$ y $\Pi_{\text{tran}}$ son las amplitudes de las ondas planas
asociadas al campo de presiones incidente, reflejada y trasmitida, respectivamente, en el dominio
de la frecuencia.

Para ver, de una forma más clara, en qué medida una superficie
es capaz de amortiguar la propagación acústica, se introduce
el coeficiente de absorción. Este coeficiente está
íntimamente relacionado con los coeficientes de reflexión ya
que se define como
\begin{equation*}
a=1-|R|^2,
\end{equation*}
lo que implica que $0\le a\le 1$. En general, la absorción tiene
en cuenta tanto la energía transmitida como la energía
disipada al propagarse la onda a través de los diferentes medios. 


Aunque no incluya en su denominación la palabra nivel, se
trata de uno de ellos y por tanto se mide en decibelios. La
pérdida de transmisión se define como el logaritmo del inverso
del coeficiente de transmisión $\tau$ asociado a la intensidad,
\begin{equation*}
\text{TL}=10\log_{10}\frac{1}{\tau}.
\end{equation*}


## Propagación de ondas planas con incidencia normal a través de dos fluidos de diferentes características

Obviamente la propagación a través de una interfaz de 
contacto que separa dos fluidos idénticos no implica reflexión alguna, puesto que
las impedancias de entrada y salida a la interfaz son iguales a 
la impedancia característica de ambos medios.
Sin embargo, si los medios poseen diferentes propiedades pueden aparecer fenómenos de reflexión.
Las ecuaciones que gobiernan de transmisión de ondas planas en este caso son
$$
\left\{\begin{array}{l}
U_1(a)=U_2(a),\\
\rho_1 c_{1}^{2}\frac{d U_1}{d p}(a)=\rho_2 c_{2}^2\frac{d U_2}{d p}(a),
\end{array}\right.
$$

```python
import numpy as np
# Constantes físicas del problem
rho1 = 1 # densidad de masa medio 1
rho2 = 2 # densidad de masa medio 2
c1 = 340 # velocidad del sonido medio 1
c2 = 480 # velocidad del medio 2
F = 1 # Amplitud incidente

def compute_TMM(rho1,rho2,c1,c2,F,omega):
    k1 = omega/c1
    k2 = omega/c2
    A = np.array([[0,1,0,0,], [1,1,-1,-1], 
                  [rho1*c1**2*(-1j*k1),rho1*c1**2*(1j*k1),-rho2*c2**2*(-1j*k2),-rho2*c2**2*(1j*k2)],
                  [0,0,1,0]])
    rhs = np.array([F,0,0,0])
    return np.linalg.solve(A, rhs)

# Barrido en frecuencia para el coeficiente de absorción
nfreq=100
freq = np.linspace(5,1000,nfreq) # Frecuencia [hz]
absorcion = np.zeros(nfreq)
for i,fi in enumerate(freq):
    Ci = compute_TMM(rho1,rho2,c1,c2,F,2*np.pi*fi)
    absorcion[i]=1-np.abs(Ci[0]/F)**2
    
# Representación gráfica
import matplotlib as mp
import matplotlib.pyplot as plt
plt.plot(freq, absorcion, label='absorción')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Absorción')
plt.legend()
plt.show()
```

## Propagación de ondas planas con incidencia normal a través de dos fluidos de las mismas características pero separados por una placa rígida

Obviamente la propagación a través de una interfaz de 
contacto que separa dos fluidos idénticos no implica reflexión alguna, puesto que
las impedancias de entrada y salida a la interfaz son iguales a 
la impedancia característica de ambos medios.
Sin embargo, la inclusión de una placa rígida entre los 
dos medios modifica esta situación (véase la Figura más bajo).
<img src="images/multicapa-placa.png" alt="Material multicapa,con una placa rígida" title="Material multicapa,con una placa rígida" width="300"/>

Las ecuaciones que gobiernan de transmisión de ondas planas en este caso son
$$
\left\{\begin{array}{l}
U_1(a)=U_2(a),\\
-\rho_1 c_{1}^{2}\frac{d U_1}{d p}(a)=-\rho_2 c_{2}^2\frac{d U_2}{d p}(a)+(\rho_{\mathrm{p}}\omega^2+i\omega s-r)U_2(a),
\end{array}\right.
$$
donde $\rho_{\mathrm{p}}$, $s$ y $r$ son la densidad de masa superficial, el coeficiente de amortiguamiento y el coeficiente de elasticidad de la placa rígida.

```python
import numpy as np
# Constantes físicas del problema
rho1 = 1 # densidad de masa medio 1
rho2 = 1 # densidad de masa medio 2
c1 = 340 # velocidad del sonido medio 1
c2 = 340 # velocidad del medio 2
F = 1 # Amplitud incidente

# Constantes físicas de la placa rígida
rhop = 0.7 # densidad de masa superficial
s = 0.2 # coeficiente de amortiguamiento
r = 10 # coeficiente de elasticidad

def compute_TMM(rho1,rho2,c1,c2,rhop,s,r,F,omega):
    k1 = omega/c1
    k2 = omega/c2
    Zw = rhop*omega**2+1j*omega*s-r # impedancia mecánica de la placa
    A = np.array([[0,1,0,0,], [1,1,-1,-1], 
                  [-rho1*c1**2*(-1j*k1),-rho1*c1**2*(1j*k1),+rho2*c2**2*(-1j*k2)-Zw,+rho2*c2**2*(1j*k2)-Zw],
                  [0,0,1,0]])
    rhs = np.array([F,0,0,0])
    return np.linalg.solve(A, rhs)

# Barrido en frecuencia para el coeficiente de transmisión
nfreq=100
freq = np.linspace(5,1000,nfreq) # Frecuencia [hz]
transmision = np.zeros(nfreq)
for i,fi in enumerate(freq):
    Ci = compute_TMM(rho1,rho2,c1,c2,rhop,s,r,F,2*np.pi*fi)
    transmision[i]=np.abs(Ci[3]/F)
    
# Representación gráfica
import matplotlib as mp
import matplotlib.pyplot as plt
plt.plot(freq, transmision, label='transmisión')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Transmisión')
plt.legend()
plt.show()
```

## Propagación de ondas planas con incidencia  normal a través de dos medios (fluidos o medios porosos) con una interfaz de contacto común y uno de ellos en contacto con una pared rígida

Si suponemos que el segundo medio es el que está en contacto 
con la pared rígida a una distancia $a$ de la interfaz común que
situamos en $p=0$ (véase la Figura más abajo), 
<img src="images/multicapa-rigida.png" alt="Material multicapa,con una placa rígida" title="Material multicapa,con una placa rígida" width="300"/>     
las condiciones de transmisión que debemos imponer en las dos interfaces son
$$
\left\{\begin{array}{l}
U_1(0)=U_2(0),\\
-\rho_1 c_{1}^{2}\frac{d U_1}{d p}(0)=-\rho_2 c_{2}^2\frac{d U_2}{d p}(0),\\
U_2(a)=0.
\end{array}\right.
$$

Si suponemos que la amplitud de la onda incidente es igual a la unidad,
el sistema lineal a resolver es
$$
\left\{\begin{array}{ccc}
C_2&=&1,\\
C_1+C_2&=&C_3+C_4, \\
Z_1 (-C_1+C_2)&=&Z_2 (-C_3+C_4),\\
C_3e^{-ik_{2}a}+C_4e^{ik_{2}a}&=&0,
\end{array}\right.
$$
donde $Z_1$ y $Z_2$ son las impedancias características del medio $1$ y $2$, respectivamente.

```python
import numpy as np
# Constantes físicas del problem
rho1 = 1 # densidad de masa medio 1
rho2 = 2 # densidad de masa medio 2
c1 = 340 # velocidad del sonido medio 1
c2 = 480 # velocidad del medio 2
F = 1 # Amplitud incidente
a = 0.1 # espesor capa 2

def compute_TMM(rho1,rho2,c1,c2,a,F,omega):
    k1 = omega/c1
    k2 = omega/c2
    A = np.array([[0,1,0,0,], [1,1,-1,-1], 
                  [rho1*c1**2*(-1j*k1),rho1*c1**2*(1j*k1),-rho2*c2**2*(-1j*k2),-rho2*c2**2*(1j*k2)],
                  [0,0,np.exp(-1j*k2*a),np.exp(1j*k2*a)]])
    rhs = np.array([F,0,0,0])
    return np.linalg.solve(A, rhs)

# Barrido en frecuencia para la impedancia de entrada
nfreq=100
freq = np.linspace(50,1000,nfreq) # Frecuencia [hz]
Zinput = np.zeros(nfreq, dtype=complex)
Zoutput = np.zeros(nfreq)
for i,fi in enumerate(freq):
    Ci = compute_TMM(rho1,rho2,c1,c2,a,F,2*np.pi*fi)
    Zinput[i]=rho1*c1*(Ci[0]-Ci[1])/(Ci[0]+Ci[1])  
    
# Representación gráfica
import matplotlib as mp
import matplotlib.pyplot as plt
plt.plot(freq, Zinput.real, label='Real')
plt.plot(freq, Zinput.imag, label='Imag')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Impedancia entrada')
plt.legend()
plt.show()
```
